```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

phNum_df = pd.read_excel(r'C:\Users\user\Desktop\hansol\ProjectB\01_2024_kr_phNum.xlsx')

columns = ['이름','인구 1만명 대비 의사수']

phNum_df = phNum_df[columns]

phNum_df = phNum_df.sort_values(by='인구 1만명 대비 의사수', ascending=False)

phNum_df = phNum_df.reset_index(drop=True)
phNum_df.index = phNum_df.index + 1

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(phNum_df['이름'], phNum_df['인구 1만명 대비 의사수'])
plt.xlabel('지역')
plt.ylabel('인구 1만명 대비 의사수')
plt.title('지역별 인구 1만명 대비 의사수')
plt.xticks(rotation=45)

# 숫자 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')


plt.show()
```
